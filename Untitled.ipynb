```python
import os
import sys
import warnings

import numpy as np
import pandas as pd
from mizani.formatters import percent_format
from plotnine import *
from datetime import datetime
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy.stats import norm
from IPython.core.display import HTML
from stargazer.stargazer import Stargazer
import statsmodels.nonparametric.kernel_regression as loess

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
plotdata =pd.read_csv("plottt.csv")

(
    ggplot(
        plotdata,
        aes(x="longitude", y="value", color="variable", linetype="variable"),
    )
    + geom_line(size=1.5)
    + scale_x_continuous(expand=(0.01, 0.01))
    + scale_y_continuous(
        expand=(0.01, 0.01), limits=(3.4, 5)
    )
    + scale_color_manual(
        values=['black', 'blue', 'red'],
        labels=["mean", "CIUP", "CILO"],
    )
    + scale_linetype_manual(
        values=["-", ":", ":"],
        labels=["mean", "CIUP", "CILO"],
    )
    + labs(x="longitude", y="price")
    + theme_bw()
    + theme(
        legend_position=(0.6, 0.2),
        legend_direction="horizontal",
        legend_title=element_blank(),
        legend_background=element_blank(),
    )
)
```
